# Image datatypes
- images = numpy arrays (unit8/16/32, float, int8/16/32)

```python
import numpy as np
from skimage.util import img_as_float
image = np.arange(0,50,10,dtype=np.uint8)
print(image.astype(np.float))
print(img_as_float(image))
```

## input types
- img_as_float, img_as_byte, img_as_uint, img_as_int

```python
from skimage.util import img_as_ubyte
image = np.array([0, 0.5, 1], dtype=float)
img_as_ubyte(image)
np.array([  0, 128, 255], dtype=np.uint8)
```

```python
# conversions can result in loss of precision
image = np.array([0, 0.5, 0.503, 1], dtype=np.float)
img_as_ubyte(image)
np.array([  0, 128, 128, 255], dtype=np.uint8)
```

```python
from skimage import data
from skimage.transform import rescale
image = data.coins()
image.dtype, image.min(), image.max(), image.shape

rescaled = rescale(image, 0.5)
print(rescaled.dtype, 
 np.round(rescaled.min(), 4), 
 np.round(rescaled.max(), 4), 
 rescaled.shape)

rescaled = rescale(image, 0.5, preserve_range=True)
print(rescaled.dtype, 
      np.round(rescaled.min()), 
      np.round(rescaled.max()), 
      rescaled.shape)
```

## output types

```python
from skimage.util import img_as_uint
from skimage.filters import sobel
import matplotlib.pyplot as plt
out = img_as_uint(sobel(image))
plt.imshow(out)
plt.show()
```
